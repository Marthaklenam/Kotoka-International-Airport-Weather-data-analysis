## Temperature conversion from Fahrenheit (°F) to Celcius (°C)
As the temperature measurement in Ghana is in celsius, our weather data has the temperature measured in Fahrenheit so we will do a conversion of our weather data to suit the  degree measurement for the country we are in.

```python
# Renaming all the columns before conversion
weather.rename(columns={'STATION': 'station', 'NAME': 'area_name', 'PRCP': 'prcp', 'TAVG': 'temp_avg', 'TMAX': 'temp_max', 'TMIN': 'temp_min'}, inplace=True)
```

```python
weather.head()
```

From the above output, columns have been renamed successfully

```python
# converting average temperature column to celcius
weather['temp_avg_°c'] = (weather['temp_avg'] - 32) * 5/9  
```

```python
# converting maximum temperature column to celcius
weather['temp_max_°c'] = (weather['temp_max'] - 32) * 5/9  
```

```python
# converting minimum temperature column to celcius
weather['temp_min_°c'] = (weather['temp_min'] - 32) * 5/9 
```

```python
# selecting only the neccessary columns needed
weather = weather[['prcp', 'temp_avg_°c', 'temp_max_°c', 'temp_min_°c']]
```

```python
weather.head()
```

From the above column, our temperature columns have been convertedto the right degree successfully.

```python
weather['temp_avg_°c'].plot(kind='hist', bins=20, color="brown", title="Temperature Distribution")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

The average temperature that occur daily from our historical data falls between 25°C to 28°C as visualized above

```python
import plotly.express as px

# Creating an interactive histogram with Plotly
fig = px.histogram(weather, x='temp_avg_°c', nbins=20, title="Temperature Distribution", color_discrete_sequence=["brown"])
fig.update_layout(
    xaxis_title="Average Temperature (°C)",
    yaxis_title="Frequency"
)
fig.show()
```

```python
fig = px.line(weather, x=weather.index, y='prcp', title="Daily Rainfall Over Time", color_discrete_sequence=["skyblue"])
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Rainfall (mm)"
)
fig.show()
```

```python
# Resample by month to get averages
monthly_avg = weather.resample('M').mean()

fig, ax1 = plt.subplots()

# Plot temperature average on left y-axis
ax1.set_xlabel("Month")
ax1.set_ylabel("Avg Temperature (°C)", color="tab:red")
ax1.plot(monthly_avg.index, monthly_avg['temp_avg_°c'], color="tab:red")
ax1.tick_params(axis="y", labelcolor="tab:red")

# Plot rainfall on right y-axis
ax2 = ax1.twinx()
ax2.set_ylabel("Avg Rainfall (mm)", color="tab:blue")
ax2.bar(monthly_avg.index, monthly_avg['prcp'], color="tab:blue", alpha=0.3)
ax2.tick_params(axis="y", labelcolor="tab:blue")

plt.title("Monthly Average Temperature and Rainfall")
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming weather is a DataFrame with a DateTime index
monthly_avg = weather.resample('M').mean()

# Create a figure
fig = go.Figure()

# Add temperature line
fig.add_trace(go.Scatter(
    x=monthly_avg.index,
    y=monthly_avg['temp_avg_°c'],
    mode='lines',
    name='Avg Temperature (°C)',
    line=dict(color='red'),
    yaxis='y1'
))

# Add rainfall bar
fig.add_trace(go.Bar(
    x=monthly_avg.index,
    y=monthly_avg['prcp'],
    name='Avg Rainfall (mm)',
    marker=dict(color='blue', opacity=0.3),
    yaxis='y2'
))

# Update layout to add y-axes and title
fig.update_layout(
    title="Monthly Average Temperature and Rainfall",
    xaxis_title="Month",
    yaxis_title="Avg Temperature (°C)",
    yaxis2=dict(
        title="Avg Rainfall (mm)",
        overlaying='y',
        side='right'
    ),
    legend=dict(x=0.1, y=1.1),
    template="plotly_white"
)

# Show the figure
fig.show()
```

```python
# Add month column based on the index to group by month
weather['Month'] = weather.index.month

# Box plot for temperature ranges grouped by month
weather.boxplot(column=['temp_min_°c', 'temp_avg_°c', 'temp_max_°c'], by='Month', grid=False)
plt.title("Monthly Temperature Range")
plt.suptitle("")  # Remove default "Boxplot grouped by Month" title
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```
